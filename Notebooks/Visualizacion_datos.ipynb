Visualizacion de Datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Scatter plot (diagramas de dispersion)
from turtle import title


x = np.arange(1,20)
y = x**2
y1= x**2 + 5*x

plt.scatter(x,y,s=10,color='red',marker='s',label='x**2 + 5*x',)
plt.scatter(x,y1,s=10, color='blue',marker='s',label='x**2')
plt.legend(title='Funciones')
plt.title('Tipos de Funciones', fontsize=20)
plt.xlabel('x', color='green')
plt.xlabel('f(x)')
plt.grid()
# plt.lim(o,5)
plt.text(5,300,'este texto',color = 'blue')
#agregar las cordenadas de los marcadores
for val_x in x:
    plt.text(val_x, val_x**2, f'({val_x}, {val_x**2}')
```

```python
#histograma
media= 20
std=3
numeros = np.random.normal(media,std, 10000)

plt.hist(numeros, bins =20, cumulative=False, density=False)
plt.xtable('Distribicion Norma')
plt.xlabel('numeros')
plt.ylabel('frecuencia')
```

```python
plt.boxplot(numeros, vert=False)
```
